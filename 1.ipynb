```python
RDspend=pd.DataFrame(data['R&D Spend'])
MSpend=pd.DataFrame(data['Marketing Spend'])
```

```python
model=smf.ols("Profit~RDspend+Administration+MSpend",data=data).fit()
```

```python
m_ms.summary()
```

```python
rsq_pro = smf.ols('Profit~RDspend+Administration+MSpend',data=data).fit().rsquared  
vif_pro = 1/(1-rsq_pro) 
rsq_rd = smf.ols('RDspend~Administration+MSpend+Profit',data=data).fit().rsquared  
vif_rd = 1/(1-rsq_rd) 
rsq_ad = smf.ols('Administration~Profit+RDspend++MSpend',data=data).fit().rsquared  
vif_ad = 1/(1-rsq_a) 
rsq_ms = smf.ols('MSpend~Profit+RDspend+Administration',data=data).fit().rsquared  
vif_ms = 1/(1-rsq_ms) 
```

```python
d1={'variance':['pro','rd','rd','ms'],'VIF':[vif_pro,vif_rd,vif_ad,vif_ms]}
VIF_frame=pd.DataFrame(d1)
VIF_frame
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') 
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-20000))
```

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "RDspend", fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
fig=sm.graphics.plot_regress_exog(model, 'Administration',fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "MSpend", fig=fig)
plt.show()
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
ig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(data)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```
